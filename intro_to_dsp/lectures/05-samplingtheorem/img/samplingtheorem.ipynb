```python
dt = 0.001
T = 20
N = int(T / dt)
time = np.arange(-2, -2 + T, dt)
M1, M2 = 50, 200
b1 = (1 / M1) * np.ones(M1)
b2 = (1 / M2) * np.ones(M2)
a = np.ones(1)
x1 = 100 * filtfilt(b1, a, np.random.randn(N), padtype='even')
x2 = 100 * filtfilt(b2, a, np.random.randn(N), padtype='even')
_x3 = 100 * filtfilt(b2, a, np.random.randn(N), padtype='even')
_jumpinx = random.randint(10000)
```

```python
_jumpsig = np.zeros(N)
_jumpsig[_jumpinx] = 10
_jumpsig[_jumpinx + 1000] = -10
_jumpsig = np.cumsum(_jumpsig)
x3 = _x3 + _jumpsig
```

```python
fig = figure(figsize=(15, 4))
ax = fig.add_subplot(131)
ax.plot(time, x1, color="tab:red")
ax.axhline(0, color="gray", linewidth=1)
ax.set_xlim(-0.55, 0.55)
ax.set_ylim(-30, 30)
ax.axis('off')

ax = fig.add_subplot(132)
ax.plot(time, 3 * x2, color="tab:blue")
ax.axhline(0, color="gray", linewidth=1)
ax.set_xlim(-0.55, 0.55)
ax.set_ylim(-30, 30)
ax.axis('off')

ax = fig.add_subplot(133)
ax.plot(time[0:_jumpinx], x3[0:_jumpinx], color="tab:green")
ax.plot(time[_jumpinx:_jumpinx+1000], x3[_jumpinx:_jumpinx+1000], color="tab:green")
ax.plot(time[_jumpinx+1000:], x3[_jumpinx+1000:], color="tab:green")
ax.axhline(0, color="gray", linewidth=1)
ax.set_xlim(2, 5)
ax.set_ylim(-30, 30)
ax.axis('off')

plt.tight_layout(w_pad=10)

fig.savefig("samp2.png", format="png", dpi=300)
fig.savefig("samp2.svg", format="svg", dpi=300)
```

```python
dt = 0.001
T = 20
N = int(T / dt)
time = np.arange(-2, -2 + T, dt)
Fs1 = 5
f0 = 0.6
f1 = 1
f2 = f1 + Fs1
f3 = f2 + 2 * Fs1
x0 = np.sin(2 * np.pi * f0 * time)
x1 = np.sin(2 * np.pi * f1 * time)
x2 = np.sin(2 * np.pi * f2 * time)
x3 = np.sin(2 * np.pi * f3 * time)
sampinx = list(range(0, N, int(1 / (dt * Fs1))))

fig = figure(figsize=(20, 8))
ax = fig.add_subplot(221)
ax.plot(time, x1, lw=5, color="tab:blue", alpha=0.3)
ax.plot(time[sampinx], x1[sampinx], 'o', color="tab:red")
ax.plot(time, x1, lw=1, color="tab:red", linestyle="--")
ax.set_xticks(np.arange(time[0], T, 1 / Fs1))
ax.set_xlim(-1, 1)
# ax.set_xlabel("Time (s)", fontsize=16)
ax.tick_params(axis = 'both', which = 'major', labelsize = 16)
ax.set_title(f"{f1}Hz Sinusoid, Fs = {Fs1}Hz", fontsize=20)
ax.grid(color='0.85', linestyle='-', linewidth=0.5)

ax = fig.add_subplot(223)
ax.plot(time, x0, lw=5, color="tab:blue", alpha=0.3)
ax.plot(time[sampinx], x0[sampinx], 'o', color="tab:red")
ax.plot(time, x0, lw=1, color="tab:red", linestyle="--")
ax.set_xticks(np.arange(time[0], T, 1 / Fs1))
ax.set_xlim(-1, 1)
ax.set_xlabel("Time (s)", fontsize=16)
ax.tick_params(axis = 'both', which = 'major', labelsize = 16)
ax.set_title(f"{f0}Hz Sinusoid, Fs = {Fs1}Hz", fontsize=20)
ax.grid(color='0.85', linestyle='-', linewidth=0.5)

ax = fig.add_subplot(222)
ax.plot(time, x2, alpha=0.6)
ax.plot(time[sampinx], x2[sampinx], 'o', color="tab:red")
ax.plot(time, x1, lw=1, color="tab:red", linestyle="--")
ax.set_xticks(np.arange(time[0], T, 1 / Fs1))
ax.set_xlim(-1, 1)
# ax.set_xlabel("Time (s)", fontsize=16)
ax.tick_params(axis = 'both', which = 'major', labelsize = 16)
ax.set_title(f"{f2}Hz Sinusoid, Fs = {Fs1}Hz", fontsize=20)
ax.grid(color='0.85', linestyle='-', linewidth=0.5)

ax = fig.add_subplot(224)
ax.plot(time, x3, alpha=0.6)
ax.plot(time[sampinx], x3[sampinx], 'o', color="tab:red")
ax.plot(time, x1, lw=1, color="tab:red", linestyle="--")
ax.set_xticks(np.arange(time[0], T, 1 / Fs1))
ax.set_xlim(-1, 1)
ax.set_xlabel("Time (s)", fontsize=16)
ax.tick_params(axis = 'both', which = 'major', labelsize = 16)
ax.set_title(f"{f3}Hz Sinusoid, Fs = {Fs1}Hz", fontsize=20)
ax.grid(color='0.85', linestyle='-', linewidth=0.5)

plt.tight_layout()
               
fig.savefig("alias.png", format="png", dpi=300)
fig.savefig("alias.eps", format="eps", dpi=300)
```
