## Modulated impulse train

```python
def x(t):
    return (0.6 + np.sin(0.1 * np.pi * t)
            + 0.3 * np.sin(0.7 * np.pi * t + 0.3 * np.pi)
            + 0.5 * np.cos(0.9 * np.pi * t))
```

```python
def Xw(w, n, Ts):
    _X = np.array([np.min((1, _w)) for _w in 4 * np.abs(w - n / Ts)])
    return 1 - _X
```

```python
fig = figure(figsize=(16, 3))
ax = fig.add_subplot(121)
ax.grid(color='0.8', linestyle='--', linewidth=0.5);
_t = np.arange(-10, 10, 0.01)
Ts = 1
_x = np.array([x(__t) for __t in _t])
ax.plot(_t, _x, lw=1, ls="--", color="tab:red", alpha=0.5)
for n in range(-30, 31):
    plt.arrow(n * Ts, -0.04, 0, x(n * Ts), width=0.025, length_includes_head=True,
              head_starts_at_zero=True, head_width=0.1, head_length=0.08, edgecolor="k", facecolor="k")
plt.axhline(0, lw=2, color="k")
ax.set_xlim(-4.5, 4.5)
ax.set_title(r"Modulated Impulse Train $x(t) \cdot \sum_{n} \delta( t - n\cdot T_s) $", fontsize=18)
ax.set_xlabel(r"Time (s)", fontsize=16)
xticks(fontsize=16)
yticks(fontsize=16)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(True)
ax.spines['bottom'].set_visible(True)

ax = fig.add_subplot(122)
ax.grid(color='0.8', linestyle='--', linewidth=0.5);
w = np.arange(-30, 30, 0.001)
w = np.arange(-50, 51, 0.01)
_Xw = np.array([Xw(w, _n, Ts) for _n in range(-20, 21)])
for i in range(_Xw.shape[0]):
    ax.plot(w, _Xw[i, :], lw=1, color="tab:red", alpha=0.4)
ax.plot(w, np.sum(_Xw, axis=0), lw=3, color="k")
ax.set_title(r"Fourier Transform of Impulse Train $X(\omega) * \frac{2\pi}{T_s} \sum_{n} \delta( \omega - n\cdot \omega_s) $", fontsize=18)
ax.set_xlabel(r"$\frac{\omega}{2\pi T_s}$ (1/sec)", fontsize=16)
xticks(fontsize=16)
yticks(fontsize=16)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(True)
ax.spines['bottom'].set_visible(True)
ax.set_xlim(-2, 2)

plt.tight_layout(w_pad=10)

fig.savefig(f"{outdir}/sampling1.svg", format="svg", dpi=300);
fig.savefig(f"{outdir}/sampling1.png", format="png", dpi=300);
```

## Non-bandlimited signal.

```python
def Xwnbl(w, n, Ts):
    _X = np.array([np.min((1, 0.25 / _w)) for _w in np.abs(w - n / Ts)])
    return _X
```

```python
fig = figure(figsize=(16, 3))
ax = fig.add_subplot(121)
ax.grid(color='0.8', linestyle='--', linewidth=0.5);
_t = np.arange(-10, 10, 0.01)
w = np.arange(-50, 51, 0.01)
Ts = 0.4
_Xw = Xwnbl(w, 0, 1)
ax.plot(w, _Xw, lw=3, color="k")
ax.set_xlim(-4.5, 4.5)
ax.set_title(r"$X(\omega)$", fontsize=18)
ax.set_xlabel(r"$\frac{\omega}{2\pi T_s}$ (1/sec)", fontsize=16)
xticks(fontsize=16)
yticks(fontsize=16)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(True)
ax.spines['bottom'].set_visible(True)

ax = fig.add_subplot(122)
ax.grid(color='0.8', linestyle='--', linewidth=0.5);
w = np.arange(-30, 30, 0.001)
w = np.arange(-50, 51, 0.01)
_Xw = np.array([Xwnbl(w, _n, Ts) for _n in range(-20, 21)])
for i in range(_Xw.shape[0]):
    ax.plot(w, _Xw[i, :], lw=1, color="tab:red", alpha=0.4)
ax.plot(w, np.sum(_Xw, axis=0), lw=3, color="k")
ax.set_title(r"Fourier Transform of Impulse Train $X(\omega) * \frac{2\pi}{T_s} \sum_{n} \delta( \omega - n\cdot \omega_s) $", fontsize=18)
ax.set_xlabel(r"$\frac{\omega}{2\pi T_s}$ (1/sec)", fontsize=16)
xticks(fontsize=16)
yticks(fontsize=16)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(True)
ax.spines['bottom'].set_visible(True)
ax.set_xlim(-4.5, 4.5)

plt.tight_layout(w_pad=10)

fig.savefig(f"{outdir}/sampling-nbl.svg", format="svg", dpi=300);
fig.savefig(f"{outdir}/sampling-nbl.png", format="png", dpi=300);
```
