## Moving average filter

```python
n = np.arange(-6, 7, 1)
h = np.zeros(len(n))
h[5:8] = 1.0 / 3


fig = figure(figsize=(8, 9))

ax = fig.add_subplot(311)
ax.grid(color='0.8', linestyle='--', linewidth=0.5);
ax.stem(n, h)
ax.set_xlim(-6, 6)
ax.set_ylim(-0.05, 0.45)
xticks(fontsize=16)
yticks(fontsize=16)
ax.set_xlabel("Time $n$", fontsize=16)
ax.set_ylabel("$h[n]$", fontsize=16)
ax.set_title("Impulse Response", fontsize=16)

ax = fig.add_subplot(312)
ax.grid(color='0.8', linestyle='--', linewidth=0.5);
omega = np.arange(-np.pi, np.pi, 0.001)
Hmag = (1 / 3) * np.abs(1 + 2 * np.cos(omega))
Harg = np.array([0 if np.abs(_w) <= 2 * np.pi / 3 else np.sign(_w) * np.pi
                 for _w in omega ])
ax.plot(omega, Hmag, lw=3)
ax.set_xlim(-np.pi, np.pi)
ax.set_ylim(-0.01, 1.1)
xticks(fontsize=16)
yticks(fontsize=16)
ax.set_xlabel(r"$\Omega$", fontsize=16)
ax.set_ylabel(r"$\vert H(\Omega) \vert$", fontsize=16)
ax.set_title("Magnitude Response", fontsize=16)

ax = fig.add_subplot(313)
ax.grid(color='0.8', linestyle='--', linewidth=0.5);
ax.plot(omega, Harg, lw=3)
ax.set_xlim(-np.pi, np.pi)
ax.set_ylim(-1.1 * np.pi, 1.1 * np.pi)
xticks(fontsize=16)
yticks(fontsize=16)
ax.set_xlabel(r"$\Omega$", fontsize=16)
ax.set_ylabel(r"$\arg \, H(\Omega)$", fontsize=16)
ax.set_title("Phase Response", fontsize=16)

plt.tight_layout(w_pad=10)

fig.savefig(f"{outdir}/movavg.svg", format="svg", dpi=300);
fig.savefig(f"{outdir}/movavg.png", format="png", dpi=300);
```

## Ideal filter

```python
w = np.arange(-np.pi, np.pi, 0.01)
Hlp = 1.0 * np.abs(w) < 1
Hhp = 1.0 * np.abs(w) > 1
Hbp = 1.0 * (np.abs(w) > 1) * (np.abs(w) < 2)
Hbs = (np.abs(w) > 2) + (np.abs(w) < 1)

fig = figure(figsize=(12, 6))
ax = fig.add_subplot(2, 2, 1)
ax.grid(color='0.8', linestyle='--', linewidth=0.5);
ax.plot(w, Hlp, lw=3, color="black")
ax.set_xlim(-np.pi, np.pi)
xticks(fontsize=16)
yticks(fontsize=16)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(True)
ax.spines['bottom'].set_visible(True)
ax.set_xlabel(r"$\Omega$", fontsize=18)
ax.set_ylabel(r"$\vert H( \Omega ) \vert$", fontsize=18)
ax.set_title("Ideal Lowpass Filter", fontsize=18)

ax = fig.add_subplot(2, 2, 2)
ax.grid(color='0.8', linestyle='--', linewidth=0.5);
ax.plot(w, Hhp, lw=3, color="black")
ax.set_xlim(-np.pi, np.pi)
xticks(fontsize=16)
yticks(fontsize=16)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(True)
ax.spines['bottom'].set_visible(True)
ax.set_xlabel(r"$\Omega$", fontsize=18)
ax.set_ylabel(r"$\vert H( \Omega ) \vert$", fontsize=18)
ax.set_title("Ideal Highpass Filter", fontsize=18)

ax = fig.add_subplot(2, 2, 3)
ax.grid(color='0.8', linestyle='--', linewidth=0.5);
ax.plot(w, Hbp, lw=3, color="black")
ax.set_xlim(-np.pi, np.pi)
xticks(fontsize=16)
yticks(fontsize=16)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(True)
ax.spines['bottom'].set_visible(True)
ax.set_xlabel(r"$\Omega$", fontsize=18)
ax.set_ylabel(r"$\vert H( \Omega ) \vert$", fontsize=18)
ax.set_title("Ideal Bandpass Filter", fontsize=18)

ax = fig.add_subplot(2, 2, 4)
ax.grid(color='0.8', linestyle='--', linewidth=0.5);
ax.plot(w, Hbs, lw=3, color="black")
ax.set_xlim(-np.pi, np.pi)
xticks(fontsize=16)
yticks(fontsize=16)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(True)
ax.spines['bottom'].set_visible(True)
ax.set_xlabel(r"$\Omega$", fontsize=18)
ax.set_ylabel(r"$\vert H( \Omega ) \vert$", fontsize=18)
ax.set_title("Ideal Bandstop Filter", fontsize=18)

plt.tight_layout()

fig.savefig(f"{outdir}/movavg.svg", format="svg", dpi=300);
fig.savefig(f"{outdir}/movavg.png", format="png", dpi=300);
```
