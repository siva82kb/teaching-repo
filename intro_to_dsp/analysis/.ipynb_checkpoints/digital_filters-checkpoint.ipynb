```python
from scipy.signal import savgol_filter
```

```python
from scipy import signal
```

```python
def get_hdn(wd, n):
    return 2 * wd * np.pi * np.sinc(wd * n)
```

```python
def get_hdn_diff(n):
    return np.array([np.cos(np.pi * _n) / _n if _n != 0 else 0 for _n in n])
```

## Noisy signal demonstration

```python
Fs = 1000
N = 10000
Nb, Wb = signal.buttord(300, 200, 3, 20, analog=False, fs=1000)
sos = signal.butter(Nb, Wb, 'high', analog=False, output='sos', fs=1000)
x = np.random.randn(N)
y = savgol_filter(x, 21, 0)
y = savgol_filter(y, 21, 0)
yclean = savgol_filter(y, 21, 0)
y = yclean + signal.sosfilt(sos, 0.1 * np.random.randn(N))
Y = np.abs(np.fft.fft(y))
f = np.arange(0, N, 1) * (Fs / N)

# Plot time domain-signal and spectrum
fig = figure(figsize=(15, 4))
ax1 = fig.add_subplot(1, 2, 1)
ax1.plot(np.arange(0, N / Fs, 1 / Fs), y, color="tab:blue")
ax1.plot(np.arange(0, N / Fs, 1 / Fs), yclean, color="k", alpha=0.5)
ax1.set_xlim(0, 500 / Fs)
ax1.set_xlabel("Time (sec)", fontsize=18)
ax1.set_ylabel("x(t)", fontsize=18)
ax1.set_title("Time domain signal", fontsize=20)
ax1.spines['right'].set_visible(False)
ax1.spines['top'].set_visible(False)
xticks(fontsize=18)
yticks(fontsize=18)

ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(f, Y)
ax2.set_xlim(0, Fs / 2)
ax2.set_xlabel("Time (sec)", fontsize=18)
ax2.set_ylabel(r"$\vert X(\omega) \vert$", fontsize=18)
ax2.set_title("Magnitude Spectrum", fontsize=20)
ax2.spines['right'].set_visible(False)
ax2.spines['top'].set_visible(False)
xticks(fontsize=18)
yticks(fontsize=18)

plt.tight_layout()

fig.savefig(f"{outdir}/example1.svg", format="svg", dpi=300);
fig.savefig(f"{outdir}/example1.png", format="png", dpi=300);
```

```python
n = np.arange(-40, 41, 1)
wd = 0.25
xd = get_hdn(wd, n)
N = 38
nshift = np.arange(-60, 61 + N // 2, 1)
w = (nshift >= 0) * (nshift < N) / (2 * np.pi)
ndtft = 4096
xw = get_hdn(wd, (nshift - N // 2)) * w
H = np.fft.fft(xw, ndtft)

fig = figure(figsize=(16, 8))
ax1 = plt.subplot2grid((2, 3), (0, 0), rowspan=1, colspan=1)
ax1.grid(color='0.8', linestyle='--', linewidth=0.5);
_w = np.linspace(0, 1, 1000)
ax1.step(_w, np.abs(_w) < wd, 'k', lw=1)
_w = np.arange(0, 2, 2 / ndtft)
ax1.step(_w, np.abs(H), 'red', lw=2, alpha=0.4)
ax1.set_xlim(0, 1)
xticks(np.arange(0, 1.25, 0.25), fontsize=18)
yticks(fontsize=18)
ax1.set_title(r"Magnitude Response", fontsize=18)
ax1.set_xlabel(r"$\Omega / \pi$", fontsize=18)

ax2 = plt.subplot2grid((2, 3), (0, 1), rowspan=1, colspan=2)
ax2.grid(color='0.8', linestyle='--', linewidth=0.5);
markerline, stemlines, baseline = ax2.stem(n, xd, linefmt='0.9', markerfmt='k.', bottom=0, use_line_collection=True)
plt.setp(stemlines, 'linewidth', 1)
plt.setp(baseline, 'linewidth', 1)
xticks(np.arange(n[0], n[-1] + 20, 20), fontsize=18)
yticks(fontsize=18)
ax2.set_xlim(n[0], n[-1])
ax2.set_title(r"$h_d[n]$", fontsize=18)
ax2.set_xlabel(r"$n$", fontsize=18)

ax3 = plt.subplot2grid((2, 3), (1, 0), rowspan=1, colspan=1)
ax3.grid(color='0.8', linestyle='--', linewidth=0.5);
_w = np.linspace(0, 1, 1000)
ax3.step(_w, np.zeros(len(_w)), 'k', lw=1)
_w = np.arange(0, 2, 2 / ndtft)
ax3.step(_w, np.unwrap(np.angle(H)), 'red', lw=2, alpha=0.4)
ax3.set_xlim(0, wd)
ax3.set_ylim(-80, 10)
xticks(np.arange(0, wd + 0.05, 0.05), fontsize=18)
yticks(fontsize=18)
ax3.set_title(r"Phase Response", fontsize=18)
ax3.set_xlabel(r"$\Omega / \pi$", fontsize=18)

ax4 = plt.subplot2grid((2, 3), (1, 1), rowspan=1, colspan=2)
ax4.grid(color='0.8', linestyle='--', linewidth=0.5);
markerline, stemlines, baseline = ax4.stem(nshift, xw, linefmt='0.9', markerfmt='r.', bottom=0, use_line_collection=True)
plt.setp(stemlines, 'linewidth', 1)
plt.setp(baseline, 'linewidth', 1)
xticks(np.arange(n[0], n[-1] + 20, 20), fontsize=18)
yticks(fontsize=18)
ax4.set_xlim(n[0], n[-1])
ax4.set_title(r"$h[n] = w[n] \cdot h_d[n]$", fontsize=18)
ax4.set_xlabel(r"$n$", fontsize=18)

plt.tight_layout()

fig.savefig(f"{outdir}/firwindow.svg", format="svg", dpi=300);
fig.savefig(f"{outdir}/firwindow.png", format="png", dpi=300);
```

```python
n = np.arange(-40, 41, 1)
wd = 0.5
xd = get_hdn(wd, n)
N = 38
nshift = np.arange(-60, 61 + N // 2, 1)
wrect = (nshift >= 0) * (nshift < N) / (2 * np.pi)
ndtft = 8192
xwrect = get_hdn(wd, (nshift - N // 2)) * wrect
Hrect = np.fft.fft(xwrect, ndtft)
whamm = np.hstack((np.zeros(61), np.hamming(N), np.zeros(60 - N // 2))) / (2 * np.pi)
xwhamm = get_hdn(wd, (nshift - N // 2)) * whamm
Hhamm = np.fft.fft(xwhamm, ndtft)

fig = figure(figsize=(15, 5))
ax1 = plt.subplot2grid((1, 2), (0, 0), rowspan=1, colspan=1)
ax1.grid(color='0.8', linestyle='--', linewidth=0.5);
_w = np.linspace(0, 2, ndtft)
ax1.step(_w, np.abs(Hrect), 'tab:red', lw=2, alpha=0.6, label="Rectangular")
ax1.step(_w, np.abs(Hhamm), 'tab:blue', lw=2, alpha=1, label="Hamming")
xticks(np.arange(0, 1.2, 0.2), fontsize=18)
ax1.set_xlim(0, 1)
yticks(fontsize=18)
ax1.set_title(r"Magnitude Response", fontsize=18)
ax1.set_xlabel(r"$\Omega / \pi$", fontsize=18)
legend(loc=1, fontsize=16, frameon=False)

ax2 = plt.subplot2grid((1, 2), (0, 1), rowspan=1, colspan=1)
ax2.grid(color='0.8', linestyle='--', linewidth=0.5);
_w = np.linspace(0, 2, ndtft)
ax2.step(_w, 20 * np.log10(np.abs(Hrect)), 'tab:red', lw=2, alpha=0.6, label="Rectangular")
ax2.step(_w, 20 * np.log10(np.abs(Hhamm)), 'tab:blue', lw=2, alpha=1, label="Hamming")
xticks(np.arange(0, 1.2, 0.2), fontsize=18)
ax2.set_xlim(0, 1)
yticks(fontsize=18)
ax2.set_title(r"Magnitude Response (dB)", fontsize=18)
ax2.set_xlabel(r"$\Omega / \pi$", fontsize=18)
legend(loc=1, fontsize=16, frameon=False)

plt.tight_layout()

fig.savefig(f"{outdir}/windowcompare.svg", format="svg", dpi=300);
fig.savefig(f"{outdir}/windowcompare.png", format="png", dpi=300);
```
