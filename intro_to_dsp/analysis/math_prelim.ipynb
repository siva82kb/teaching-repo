## Linear function

```python
x = np.arange(-5, 5, 0.01)
k1, k2 = 0.5, -1.2
y1, y2 = k1 * x, k2 * x

fig = figure(figsize=(8, 6))
ax = fig.add_subplot(111)
ax.plot(x, y1, label=f"$y = {k1}\\cdot x$")
ax.plot(x, y2, label=f"$y = {k2}\\cdot x$")
ylim(-2.5, 2.5)
xlabel("x", fontsize=18)
ylabel("y = f(x)", fontsize=18)
title("Linear function", fontsize=18)
legend(loc=0, fontsize=15)
grid(color='0.8', linestyle='--', linewidth=0.5);

fig.savefig(f"{outdir}/linfunc.svg", format="svg", dpi=300);
fig.savefig(f"{outdir}/linfunc.png", format="png", dpi=300);
```

### Quardatic function

```python
x = np.arange(-5, 5, 0.01)
a, b, c = 0.2, -0.1, -2.0
y = a * x ** 2 + b * x + c

fig = figure(figsize=(8, 6))
ax = fig.add_subplot(111)
ax.plot(x, y)
# ylim(-2.5, 2.5)
xlabel("x", fontsize=18)
ylabel("y = f(x)", fontsize=18)
title("Quadratic function", fontsize=18)
legend(loc=0, fontsize=15)
grid(color='0.8', linestyle='--', linewidth=0.5);

fig.savefig(f"{outdir}/quadfunc.svg", format="svg", dpi=300);
fig.savefig(f"{outdir}/quadfunc.png", format="png", dpi=300);
```

## Exponential function

```python
x = np.arange(-2, 2, 0.01)
a1, k1 = 1.0, 1.0
a2, k2 = 2.0, -0.5
y1, y2 = a1 * np.exp(k1 * x), a2 * np.exp(k2 * x)


fig = figure(figsize=(8, 6))
ax = fig.add_subplot(111)
ax.plot(x, y1, label=f"$y = {a1}e^{{{k1}x}}$")
ax.plot(x, y2, label=f"$y = {a2}e^{{{k2}x}}$")
xlabel("x", fontsize=18)
ylabel("y = f(x)", fontsize=18)
title("Exponential function", fontsize=18)
legend(loc=0, fontsize=15)
grid(color='0.8', linestyle='--', linewidth=0.5);

fig.savefig(f"{outdir}/expfunc.svg", format="svg", dpi=300);
fig.savefig(f"{outdir}/expfunc.png", format="png", dpi=300);
```

### Sin and cosine function

```python
x = np.arange(-5, 5, 0.01)
ys, yc = np.sin(x), np.cos(x)

fig = figure(figsize=(8, 6))
ax = fig.add_subplot(111)
ax.plot(x, ys, label=f"$y_s = \sin(x)$")
ax.plot(x, yc, label=f"$y_c = \cos(x)$")
xlabel("x", fontsize=18)
ylabel("y = f(x)", fontsize=18)
title("Cosine and Sine functions", fontsize=18)
legend(loc=0, fontsize=15)
grid(color='0.8', linestyle='--', linewidth=0.5);

fig.savefig(f"{outdir}/cosinfunc.svg", format="svg", dpi=300);
fig.savefig(f"{outdir}/cosinfunc.png", format="png", dpi=300);
```
